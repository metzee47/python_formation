```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://fr.m.wikipedia.org/wiki/Wikip√©dia:Accueil_principal'
```

```python
# Fonction permettant d'afficher et de parser le contenu html de la page 

def get_n_parse(url):
    page = requests.get(url)
    content = BeautifulSoup(page.content,'html.parser')
    return content

# Fontion qui permet d'afficher le titre de l'article de la page d'accueil de wikipedia

def get_article_title(url):
    page = requests.get(url)
    content = BeautifulSoup(page.content,'html.parser')
    title = content.find('b').text
    return title
    
get_article_title(url)
```

```python
# Une fonction qui nous permet de recueillir le titre et le contenu du sommaire de la page python programming langage dans un dictionnaire de donnees

def get_title_value():
    link = 'https://en.wikipedia.org/wiki/Python_(programming_language)'
    page = requests.get(link)
    soup = BeautifulSoup(page.text,'html.parser')
    tbody = soup.find('tbody')
    title = tbody.find_all('th')
    t_value = tbody.find_all('td',{'class':['infobox-data','infobox-full-data']})
    my_dict ={'Titres':[],'Elements':[]}
    for i in range(len(t_value)):
        if t_value[i].text != '':
            my_dict['Elements'].append(t_value[i].text)
    for j in range(len(title)):
            my_dict['Titres'].append(title[j].get_text())
    return my_dict
            
get_title_value()
```

```python
# Une fonction qui permet de recuperer tous les liens d'une page wikipedia 

def get_page_links(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.text,'html.parser')
    link1 = soup.find_all('link')
    link2 = soup.find_all('a')
    links = link1 + link2
    for link in links:
        if link['href'][0] != 'h':
            print('https://fr.wikipedia.org' + link['href'])
        else:
            print(link['href'])
get_page_links(url)
```

```python
# Regroupons les precedentes fonctions dans une fonction principale

def general_function(url):
    getNParse = get_n_parse(url)
    getATitle = get_article_title(url)
    print (f"Tous les liens de la page ci-dessus: \n{get_page_links(url)}\n\n\nLe contenu HTML de la page: \n{getNParse}\n\n\nL'article de la page: \n{getATitle}\n\n")
```

```python
# Testons la fonction dans une page wikipedia de notre choix

url = 'https://en.wikipedia.org/wiki/Main_Page'
general_function(url)
```
