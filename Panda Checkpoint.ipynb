```python
import pandas as pd
import numpy as np

# Dictionnaire qui va nous servir de donnees du fichier excel
exam_data = {
    'name':['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}
# initialisation des indices de notre table de fichier
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# creation du dataframe a partir de notre dictionnaire de donnees (exam_data)
class_df = pd.DataFrame(exam_data, labels)
class_df
```

```python
# affichons les trois premiers lignes de notre dataframe

class_df.head(3)
class_df
```

```python
# On va supprimer les lignes ayant comme valeur NAN

drop_df = class_df.dropna(axis=0)
drop_df
```

```python
# Supprimons la colonne nom et score du dataframe

new_df = class_df.drop(['name','score'],axis=1)
new_df
```

```python
# Ajoutons une nouvelle ligne d'indice k avec ses propres donnees

new_class_df = pd.concat([class_df,pd.DataFrame({
    'name':'Suresh',
    'score':15.5,
    'attempts':1,
    'qualify':'yes'
}, list('k'))])
new_class_df
```

```python
# Un programme qui va supprimer la colonne 'attempts' du DataFrame.

new_class_df = new_class_df.drop('attempts', axis=1)
new_class_df
```

```python
# Ajoutons la colonne succes en relation directe avec la colonne score

df_index = list('abcdefghijk')
df_size = len(new_class_df)

value = [1 if new_class_df.loc[i,'score'] > 10 else 0 for i,j in zip(df_index,range(0,df_size))]

new_class_df['success'] = value
new_class_df
```

```python
# Exportons notre DataFrame vers un fichier csv

new_class_df.to_csv('CSV\my_data.csv')
```
